## Data Visualisation

```python
Researchers_All_Years_Male = Researchers_df[((Researchers_df["Region"] == "State") & (Researchers_df["Sex"] == "Male" ) & (Researchers_df["Type of Researcher"] == "All research and development staff"))]
Researchers_All_Years_Male

# All male R&D staff in the state for 2017, 2019 and 2021
# This dataset is used in Graph no.1
```

```python
Researchers_All_Years_Female = Researchers_df[((Researchers_df["Region"] == "State") & (Researchers_df["Sex"] == "Female" ) & (Researchers_df["Type of Researcher"] == "All research and development staff"))]
Researchers_All_Years_Female

# All female R&D staff in the state for 2017, 2019 and 2021
# This dataset is used in Graph no.1
```

```python
Researchers_All_Years_Both = Researchers_df[((Researchers_df["Region"] == "State") & (Researchers_df["Sex"] == "Both sexes" ) & (Researchers_df["Type of Researcher"] == "All research and development staff"))]
Researchers_All_Years_Both
# Total R&D staff in the state for 2017, 2019 and 2021
# This dataset is used in Graph no.1
```

```python
import pandas as pd
import matplotlib.pyplot as plt 

plt.plot(Researchers_All_Years_Male["Year"], Researchers_All_Years_Male["VALUE"], color = "blue", marker = "o", label = "Male", linestyle = "--")
plt.plot(Researchers_All_Years_Female["Year"], Researchers_All_Years_Female["VALUE"], color = "red", marker = "o", label = "Female", linestyle = "--")
plt.plot(Researchers_All_Years_Both["Year"], Researchers_All_Years_Both["VALUE"], color = "green", marker = "o", label ="Both", linestyle = "--")
plt.xlim(2016, 2022)
plt.title(" Graph 1: All Research and Development Staff Across Years", fontsize = 20)
plt.xlabel ("Year")
plt.ylabel ("Number of Researchers")
plt.rcParams["figure.figsize"] = (16,8)
plt.grid(True)
plt.legend(loc="upper left")
plt.show

# a lineplot which shows the number of Male and Female R&D personnel in the state across 2017, 2019 and 2021
```

```python
Researchers_2021 = Researchers_df[((Researchers_df["Region"] == "State") & (Researchers_df["Sex"] == "Male" ) & (Researchers_df["Year"] == 2021))]
Researchers_2021

# The number of male researchers in the state for 2021 by type including the total number of R&D staff
# This dataset is used in Graph no.2
# There are considerably more male R&D staff than female R&D staff and male researchers are increasing at a faster rate in the period in which the data covers
```

```python
import matplotlib.pyplot as plt 
import pandas as pd

colors = ["green", "blue", "red", "brown", "pink"]
plt.bar(Researchers_2021["Type of Researcher"], Researchers_2021["VALUE"], color=colors)
plt.title("Graph 2: Number of Male Researchers by Type (2021)", fontsize = 20)
plt.xlabel("Type of Researcher", fontsize = 12)
plt.ylabel("Number", fontsize = 12)
plt.grid(True)
plt.xticks(fontsize = 10)
plt.rcParams["figure.figsize"] = (16,8)
plt.show

# added in a grid to make the chart easier to read
# shows the total number of researchers
# changed the width and length of the graph so labels could be read easily 
# increased the size of the x and y axes labels to distinguish them from graph labels
# The number of male R&D staff from highest to lowest is the following: Other Researchers, Technicians, Support Staff and PhD qualified researchers
```

```python
Researchers_2021_Type = Researchers_2021.drop(125)
Researchers_2021_Type

# dropped the All research and development staff using the row index
# Male researchers in the state broken down by type excluding total number of R&D staff
# This dataset is used in Graph no.3
```
